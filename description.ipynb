# Life as Time and Data: Having an honest relationship with my phone

In 2021, I realized I was spending too much time on my phone. Spoilers, after four years, this is still true. The phone can be a soul-sucking, creativity-quashing blight or an invaluable tool for connection, curiosity, and data analysis. While there are many apps to track phone time, I fundementally disagreed with how these apps categorized my time. A video call could easily be miscategorized as social media time, something I was trying to decrease, when in reality it was valuable social time spend with international friends.

I came to the opinon that I'm the only one qualified to judge my time, and started using an app (aTimelogger) to differentiate my time on my phone into difference categories. It wasn't too long before I started tracking everything. The rules are simple:

1. Select relevant, recognizable categories (Less than 20 seems to be the sweet spot for me)
2. Each period is labeled only one type of category
3. If more than one type of category is in play (exercising with friends), the time is categorized by the  activity's most relevant category.
4. Periods can be corrected after the fact, and best guesses can be used if data wasn't actually taken.
5. I don't track sleep.

Several people in my friend group find all these rules controversial. While this was orginally all about me, my friends are a large part of the data. The data is entitled to expressing itself; however, I have not changed the rules because it would make the data tracking/analysis harder. Regardless, the experient has gone on long enough, and its time to start publishing results.
